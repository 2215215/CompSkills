```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def y(x):
    return np.exp(-x) * np.sin(2*x + np.pi/3)

# Generate some x values
x = np.linspace(0, 10, 100)

# Calculate the y values
y_values = y(x)

# Plot the function
plt.plot(x, y_values)

# Add a title and axis labels
plt.title("y = e^(-x) * sin(2*x + pi/3)")
plt.xlabel("x")
plt.ylabel("y")

# Show the plot
plt.show()
```
